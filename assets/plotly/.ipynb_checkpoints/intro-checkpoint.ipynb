```python
import numpy as np
import plotly.graph_objects as go

def f(x):
    return -np.exp(-((x - 0.37)**2 / 2.5)) + 1.5

def f_prime(x):
    return (2 / 2.5) * (x - 0.37) * np.exp(-((x-0.37)**2/ 2.5))

t_values = np.array([0.0, 0.126, 0.239, 0.352, 0.504, 0.695, 1.0])

c = -3.0

def euler_method(t_values, x_start, c):
    t_points = t_values[t_values >= x_start]
    if len(t_points) == 0:
        return np.array([]), np.array([])
    
    # 初值
    y0 = f(x_start)
    y_euler = [y0]
    for i in range(len(t_points)-1):
        dt = t_points[i+1] - t_points[i]
        penalty = -c*(y_euler[-1] - f(t_points[i]))
        dydt = f_prime(t_points[i]) + penalty
        y_next = y_euler[-1] + dydt * dt
        y_euler.append(y_next)
        
    return t_points, np.array(y_euler)

fig = go.Figure()

x_exact = np.linspace(0, 1, 200)
y_exact = f(x_exact)
fig.add_trace(go.Scatter(
    x=x_exact, 
    y=y_exact, 
    mode='lines', 
    line=dict(color='red', width=2),
    name='Exact solution',
    showlegend=False,
))

# 在 exact solution 上加红点表示每个起点
start_points = [0.695, 0.504, 0.352, 0.239, 0.126, 0.0]
colors = ['black']*len(start_points)

# 在 exact solution 上添加对应起点的红点
x_starts = start_points
y_starts = [f(sp) for sp in x_starts]
fig.add_trace(go.Scatter(
    x=x_starts,
    y=y_starts,
    mode='markers',
    marker=dict(color='red', size=6),
    showlegend=False
))

# 绘制 Euler 线及箭头
for i, sp in enumerate(start_points):
    tx, ty = euler_method(t_values, sp, c)
    fig.add_trace(go.Scatter(
        x=tx, 
        y=ty, 
        mode='lines',
        line=dict(color=colors[i], width=1.5),
        showlegend=False
    ))
    for j in range(len(tx)-1):
        fig.add_annotation(
            x=tx[j+1], 
            y=ty[j+1],
            ax=tx[j],
            ay=ty[j],
            xref="x", yref="y", axref="x", ayref="y",
            arrowhead=3,
            arrowsize=2,
            arrowwidth=1,
            arrowcolor="black",
            showarrow=True,
            text=""
        )

# y_轴上只显示 y_0，不显示数值刻度
fig.update_yaxes(
    showline=True,
    linecolor='black',
    mirror=True,
    showticklabels=False,  # 不显示刻度值
    ticks=""                # 不显示刻度线
)
# 不要 y 轴标题
fig.update_yaxes(title="")

# 在 y=f(0) 处添加 y_0 标识
y0_val = f(0.0)
fig.add_annotation(
    xref='paper', yref='y',
    x=-0.04, y=y0_val,
    text="$y_0$",
    showarrow=False,
    font=dict(size=11)
)

fig.update_layout(
    xaxis=dict(
        title='t',       # 横轴改为 t
        zeroline=False,
        showline=True,
        linecolor='black',
        mirror=True,
        range=[0,1.01]   # 范围稍微超出一点 1 显示箭头
    ),
    plot_bgcolor='white',
    width=450,
    height=300,
    margin=dict(l=40, r=10, t=10, b=0),
)

fig.show()

fig.write_html('plotly_chart.html', include_mathjax='cdn')
```
